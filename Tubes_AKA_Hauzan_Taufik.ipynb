## Import Libary

```python
import time
import random
import string
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

## Import Data Dummy

```python
# Mengasumsikan file '/content/daftar_kontak_dummy.csv' ada dan memiliki kolom 'Name'
try:
    # Membaca file CSV dan hanya mengambil kolom 'Name'
    dummy_names = pd.read_csv('/content/daftar_kontak_dummy.csv', usecols=["Name"])
    # Mengambil nama sebagai array terurut
    name_array = sorted(dummy_names["Name"].tolist())
except FileNotFoundError:
    # Jika file tidak ditemukan, buat array kosong
    name_array = []
```

## Algoritma Linear_Search(rekursif) dan Binary_Search(iteratif)

```python
# Algoritma Iteratif Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Algoritma Rekursif Linear Search
def linear_search_recursive(arr, target, index=0):
    if index >= len(arr):
        return -1
    if arr[index] == target:
        return index
    return linear_search_recursive(arr, target, index + 1)
```

```python
# Uji coba performa
if name_array:
    search_target = random.choice(name_array)  # Pilih target secara acak dari data
else:
    search_target = None  # Jika array kosong

if search_target:
    # Waktu Binary Search
    start_binary = time.time()
    result_binary = binary_search(name_array, search_target)
    end_binary = time.time()

    # Waktu Linear Search
    start_linear = time.time()
    result_linear = linear_search_recursive(name_array, search_target)
    end_linear = time.time()

    # Hasil
    print(f"Target: {search_target}")
    print(f"Binary Search Result: {result_binary}, Time: {end_binary - start_binary:.10f} seconds")
    print(f"Linear Search Result: {result_linear}, Time: {end_linear - start_linear:.10f} seconds")
else:
    print("Array 'name_array' kosong, tidak ada data untuk dicari.")
```

```python
import matplotlib.pyplot as plt

# Pastikan 'name_array' digunakan dan valid
if name_array:
    search_target = random.choice(name_array)  # Pilih target secara acak dari data

    # Waktu Binary Search
    start_binary = time.time()
    result_binary = binary_search(name_array, search_target)
    end_binary = time.time()

    # Waktu Linear Search
    start_linear = time.time()
    result_linear = linear_search_recursive(name_array, search_target)
    end_linear = time.time()

    # Hasil
    print(f"Target: {search_target}")
    print(f"Binary Search Result: {result_binary}, Time: {end_binary - start_binary:.10f} seconds")
    print(f"Linear Search Result: {result_linear}, Time: {end_linear - start_linear:.10f} seconds")

    # Plot Grafik
    methods = ["Binary Search", "Linear Search"]
    times = [end_binary - start_binary, end_linear - start_linear]

    plt.figure(figsize=(10, 6))
    plt.bar(methods, times)
    plt.title("Comparison of Search Algorithms")
    plt.ylabel("Time (seconds)")
    plt.xlabel("Search Method")
    plt.show()
else:
    print("Array 'name_array' kosong, tidak ada data untuk diuji.")
```

```python
# Meningkatkan batas rekursi sementara
sys.setrecursionlimit(10000)

# Pastikan array 'name_array' terisi sebelum pengujian
if name_array:
    # Uji coba performa untuk posisi target di awal, tengah, dan akhir
    positions = [0, len(name_array) // 2, len(name_array) - 1]
    position_labels = ["Awal", "Tengah", "Akhir"]
    results_binary = []
    results_linear = []

    for pos in positions:
        search_target = name_array[pos]

        # Waktu Binary Search
        start_binary = time.time()
        binary_search(name_array, search_target)
        end_binary = time.time()

        # Waktu Linear Search
        start_linear = time.time()
        linear_search_recursive(name_array, search_target)
        end_linear = time.time()

        results_binary.append(end_binary - start_binary)
        results_linear.append(end_linear - start_linear)

    # Plot Grafik
    x = range(len(position_labels))
    plt.figure(figsize=(10, 6))
    plt.plot(x, results_binary, label="Binary Search", marker="o")
    plt.plot(x, results_linear, label="Linear Search", marker="o")
    plt.xticks(x, position_labels)
    plt.title("Comparison of Search Algorithms Based on Target Position")
    plt.ylabel("Time (seconds)")
    plt.xlabel("Target Position")
    plt.legend()
    plt.show()
else:
    print("Array 'name_array' kosong, tidak ada data untuk diuji.")
```

```python
# Pastikan data sudah ada di 'name_array'
if name_array:
    input_sizes = [300, 500, 700, 1000]
    binary_times = []
    linear_times = []

    for size in input_sizes:
        # Ambil subset data dari array yang sudah terurut
        subset_array = name_array[:size]

        # Target diambil dari posisi tengah subset
        search_target = subset_array[len(subset_array) // 2]

        # Waktu Binary Search
        start_binary = time.time()
        binary_search(subset_array, search_target)
        end_binary = time.time()
        binary_times.append(end_binary - start_binary)

        # Waktu Linear Search
        start_linear = time.time()
        linear_search_recursive(subset_array, search_target)
        end_linear = time.time()
        linear_times.append(end_linear - start_linear)

    # Plot grafik perbandingan
    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, binary_times, label="Binary Search", marker="o", linestyle="--")
    plt.plot(input_sizes, linear_times, label="Linear Search", marker="o", linestyle="-")
    plt.title("Time Complexity Comparison: Binary Search vs Linear Search")
    plt.xlabel("Input Size (Number of Elements)")
    plt.ylabel("Time (seconds)")
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("Array 'name_array' kosong, tidak ada data untuk diuji.")
```
